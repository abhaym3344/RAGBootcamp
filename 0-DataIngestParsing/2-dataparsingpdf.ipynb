## Handling PDF Challenges

### Purpose of This Section PDF are notoriously difficult to parse because they:
- Store text in complex ways (not just simple text)
- Can have formatting issues
- May contain scanned images (requiring OCR)
- Often have extraction artifacts

```python
# Example of raw PDF extraction

raw_pdf_text = """ Company Financial Report

The ﬁnancial perƒormance for ﬁscal year 2024 shows signiﬁcant growth in proﬁtability.

Revenue increased by 25%, reﬂecting strong marƒket demand and improved saƖes strategies.

The company’s eﬃciency improved due to workﬂow optimiƶation and reduced operaƭional costs.

Key highlights:
• Gross proﬁt margin expanded from 32% to 38%.
• Operating income rose signiﬁcantly due to cost-eﬀective measures.
• Cash ﬂow remained stable despite increased investments.

Addiƭional notes:
The organizaƭion conƭinues to leverage automaƭion tools to improve producƭivity. However, some ineﬃciencies perƭain in legacy sysƭems.

Page 1 of 10
"""
```

```python
import re

def clean_text(text: str) -> str:

    # Fix common ligatures
    ligatures = {
        "ƭ": "t",
        "ƶ": "z",
        "Ɩ":"l",
        "’": "'",
        "ƒ": "f",
        "ﬁ": "fi",
        "ﬂ": "fl",
        "ﬀ": "ff",
        "ﬃ": "ffi",
        "ﬄ": "ffl",
        "ﬅ": "ft",
        "ﬆ": "st",
    }
    for lig, replacement in ligatures.items():
        text = text.replace(lig, replacement)

    # Remove weird/control characters (except basic punctuation)
    text = re.sub(r"[^\x00-\x7F]+", " ", text)

    # Fix broken spacing (PDFs often insert random spaces)
    text = re.sub(r"\s+", " ", text)

    # Fix spacing before punctuation
    text = re.sub(r"\s+([.,!?;:])", r"\1", text)

    # Optional: fix spaced-out words like "f i n a n c e"
    text = re.sub(r"\b(?:[a-zA-Z]\s){2,}[a-zA-Z]\b",
                  lambda m: m.group(0).replace(" ", ""),
                  text)

    return text.strip()

cleaned_text = clean_text(raw_pdf_text)

print("Raw PDF Text:")
print(repr(raw_pdf_text))
print("\nCleaned PDF Text:")
print(repr(cleaned_text)) 
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
import re
from typing import List
class SmartPDFProcessor:
    """Advanced PDF Processing with error handling"""
    def __init__(self, chunk_size=1000, chunk_overlap=100):
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=self.chunk_size,
            chunk_overlap=self.chunk_overlap,
            separators=[" "]
        )

    def process_pdf(self, pdf_path: str) -> List[Document]:
        """Process PDF with smart chunking and metadata enhancement"""

        # Load PDF
        loader = PyPDFLoader(pdf_path)
        pages = loader.load()

        ## Process each page

        processed_chunks = []

        for page_num, page in enumerate(pages):
            ## clean text
            cleaned_text = self._clean_text(page.page_content)

            # skip nearly empty pages
            if len(cleaned_text) < 50:
                print(f"Warning: Page {page_num} has very little text after cleaning. Skipping.")
                continue

            # Create chhunks with enhanced metadata
            chunks = self.text_splitter.create_documents(
                texts=[cleaned_text],
                metadatas=[{
                    **page.metadata,
                    "page_number": page_num+1,
                    "totak_pages": len(pages),
                    "chunk_method": "smart_pdf_processor",
                    "char_count": len(cleaned_text)
                }]
            )
            processed_chunks.extend(chunks)
        return processed_chunks
    
    def _clean_text(self, text: str) -> str:
        """Clean text by fixing common PDF extraction issues"""
        # Fix common ligatures
        ligatures = {
            "ƭ": "t",
            "ƶ": "z",
            "Ɩ":"l",
            "’": "'",
            "ƒ": "f",
            "ﬁ": "fi",
            "ﬂ": "fl",
            "ﬀ": "ff",
            "ﬃ": "ffi",
            "ﬄ": "ffl",
            "ﬅ": "ft",
            "ﬆ": "st",
        }
        for lig, replacement in ligatures.items():
            text = text.replace(lig, replacement)

        # Remove weird/control characters (except basic punctuation)
        text = re.sub(r"[^\x00-\x7F]+", " ", text)

        # Fix broken spacing (PDFs often insert random spaces)
        text = re.sub(r"\s+", " ", text)

        # Fix spacing before punctuation
        text = re.sub(r"\s+([.,!?;:])", r"\1", text)

        # Optional: fix spaced-out words like "f i n a n c e"
        text = re.sub(r"\b(?:[a-zA-Z]\s){2,}[a-zA-Z]\b",
                    lambda m: m.group(0).replace(" ", ""),
                    text)

        return text.strip()
```
