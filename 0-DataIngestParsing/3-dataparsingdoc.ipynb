## Word Document Processing

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
## Mehtod 1: Using Docx2txtLoader
print("Using Docx2txtLoader:\n")
try:
    doc_loader = Docx2txtLoader("data/word_files/proposal.docx")
    documents = doc_loader.load()
    for doc in documents:
        print(f"Loaded {len(documents)} document(s) with Docx2txtLoader.")
        print(doc.metadata)  # Print metadata
        print(doc.page_content[:200])  # Print the first 200 characters of the content
except Exception as e:
    print(f"Error loading document with Docx2txtLoader: {e}")
```

```python
## Method 2: Using UnstructuredWordDocumentLoader
print("\nUsing UnstructuredWordDocumentLoader:\n")

try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/word_files/proposal.docx", mode="elements") # Use "elements" mode to get more structured output
    unstructured_docs = unstructured_loader.load()
    print(f"Loaded {len(unstructured_docs)} element(s) with UnstructuredWordDocumentLoader.")
    for i, doc in enumerate(unstructured_docs[:5]):  # Print the first 5 documents for brevity
        print(f"\nElement {i+1}:")
        print(f"Type: {doc.metadata.get('category', 'Unknown')}")
        print(doc.page_content[:100])  # Print the first 100 characters of the content
except Exception as e:
    print(f"Error loading document with UnstructuredWordDocumentLoader: {e}")
```
